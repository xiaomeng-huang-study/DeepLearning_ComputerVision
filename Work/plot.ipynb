```python
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline
import pandas as pd
import os
import numpy as np
```

# Training

## Accuracy

```python
phase = "train"
metric = "Accuracy"
archs = ["ResNet18", "ResNet50"]

log_path = "logs"
dfs = []
for arch in archs:
    df = pd.read_csv(os.path.join(log_path, arch, phase, metric + ".csv"))
    dfs.append(df)
```

```python
fig, ax = plt.subplots(figsize=(9.6, 6), dpi=300)

labels = archs

for counter_df in range(len(dfs)):
    df = dfs[counter_df]

    x = df.loc[:, ["Step"]].astype(np.int64).to_numpy()
    y = df.loc[:, ["Value"]].astype(np.float64).to_numpy()

    x = np.reshape(x, (x.shape[0],))
    y = np.reshape(y, (y.shape[0],))

    x_smooth = np.linspace(x.min(), x.max(), 500) 
    spl = make_interp_spline(x, y)
    y_smooth = spl(x_smooth)
    ax.plot(x_smooth, y_smooth, label= labels[counter_df])
    

ax.set_xlabel("Epochs", loc= "right", size= 15)
# ax.set_xticks(np.arange(-1, 200 + 1, step= 50), np.arange(-1, 200 + 1, step= 50).tolist())
# ax.set_xlim(left= 0, right= 200)

ax.set_ylabel(metric, loc= "top", size= 15)
# ax.set_yticks(np.arange(0, 1 + 0.1, step= 0.1), np.arange(0, 1 + 0.1, step= 0.1).tolist())
# ax.set_ylim(bottom=0, top= 1)

ax.grid(c= "#E7E7E7", linestyle='dashed', zorder= 1)
ax.set_title(f"{metric} over Epochs", size= 20)
ax.legend()

plt.show()
```
